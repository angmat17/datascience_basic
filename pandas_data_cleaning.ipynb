```python
# pandas 라이브러리를 임포트합니다.
import pandas as pd

# 예제 데이터프레임 생성
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 34, 29, 28],
        'Income': [100, 150, 20, 30],
        'City': ['New York', 'Paris', 'Berlin', 'London']}
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])

df
```

## 데이터프레임에서 컬럼 선택하기

```python
# 'A' 컬럼만을 선택하여 출력합니다.
df['Age']
```

## `unique()` 함수 사용하기

```python
# 'A' 컬럼의 고유값을 찾아 출력합니다.
df['Age'].unique()
```

## `loc`와 `iloc`의 사용법

```python
# loc vs iloc
# - loc는 라벨 기반으로 데이터를 선택하는 데 사용됩니다. 범위의 끝이 포함됩니다.
# - iloc는 위치 기반으로 데이터를 선택하는 데 사용됩니다. 범위의 끝이 포함되지 않습니다.
# 데이터 분석에서 loc과 iloc의 선택은 데이터 선택의 필요성과 개인의 선호에 따라 달라집니다.

# loc 사용 예시: 라벨 기반 선택
# 인덱스 라벨 'b'에서 'd'까지 모든 컬럼 선택 (끝 포함)
print("loc을 사용한 라벨 기반의 행 선택:")
print(df.loc['b':'d'])

# loc을 사용하여 'Name'과 'Age' 컬럼만 선택
print("\nloc을 사용한 라벨 기반의 컬럼 선택:")
print(df.loc[:, ['Name', 'Age']])

# iloc 사용 예시: 위치 기반 선택
# 0번째부터 2번째 인덱스 행 선택 (2번째는 포함되지 않음)
print("\niloc을 사용한 위치 기반의 행 선택:")
print(df.iloc[0:2])

# 첫 번째 행의 첫 번째와 두 번째 컬럼 선택 (컬럼도 위치 기반으로 선택)
print("\niloc을 사용한 위치 기반의 컬럼 선택:")
print(df.iloc[0, 0:2])
```

## 조건에 맞는 행 필터링

```python
df[df['Age'] == 28]
```

## 상관관계 분석

```python
# 두 컬럼 사이 상관관계를 계산합니다.
correlation = df['Age'].corr(df['Income'])
print(correlation)
```
